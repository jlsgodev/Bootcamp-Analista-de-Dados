```python
#%%

# 1 - Crie um DataFrame com as colunas nome, idade, altura, peso e sexo usando listas.
import pandas as pd                                                          # Importa a biblioteca pandas
nome = ['Daiane','Isabela','Enzo','Isabelly','Heitor','Henrique','Suelyn',]     # Cria uma lista com os nomes
idade = [30,13,9,10,6,1,33]
altura = [1.67,1.23,1.30,1.58,1.15,0.60,1.66]
peso = [69,33,40,41,28,7,66]
sexo = ['F','F','M','F','M','M','F']
df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),      # Cria o DataFrame
columns=['nome','idade','altura','peso','sexo'])         # Define os nomes das colunas

# 2 - Exiba o DataFrame criado.
#display(df)                                             # Exibe o DataFrame


# 3 - Selecione e exiba a coluna "idade" do DataFrame.
#display(df['idade'])                                  # Exibe a coluna "idade"


# 4 - Selecione e exiba a linha com índice 2 do DataFrame.
#df.loc[2]              # vai procurar a linha que tem o índice 2 (não a terceira linha necessariamente).


# 5 - Selecione e exiba a linha com rótulo 2 do DataFrame.
#df.iloc[2]         # vai pegar a linha na posição 2, que é a terceira linha do DataFrame, independentemente do valor do índice.



# 6 - Selecione e exiba o valor da célula na linha de rótulo 3, coluna "altura".
#df.loc[3,'altura']     # Seleciona somente o valor da linha 3 e coluna "altura".


# 7 - Renomeie as colunas "sexo" para "genero" e "peso" para "massa".
df.rename(columns={'sexo':'genero','peso':'massa'},inplace=True)  # Renomeia as colunas
#display(df)


# 8 - Selecione e exiba as colunas "nome" e "altura" do DataFrame.
df[['nome','altura']].groupby('nome')    # Agrupa as colunas "nome" e "altura"
#display(df[['nome', 'altura']])     # Exibe as colunas "nome" e "altura"

# 9 - Filtre e exiba todas as linhas onde a idade é maior que 40.
df[df['idade'] > 40]        # Filtra as linhas onde a idade é maior que 40
#display(df[df['idade'] > 40])       # Exibe as linhas onde a idade é maior que 40

# 10 - Adicione uma nova coluna chamada "IMC" ao DataFrame, calculando o Índice de Massa Corporal.
df['IMC'] = df['massa'] / (df['altura'] ** 2)   # Adiciona a coluna "IMC" já calculando o IMC
df['IMC'] = df['IMC'].round(1)                  # Arredonda o IMC
#display(df)


# 11 - Ordene o DataFrame pela coluna "altura" em ordem crescente.
#df.sort_values(by='altura', inplace=True)       # .sort_values() ordena o DataFrame pela coluna "altura". inplace=True faz a alteração no DataFrame original.
#display(df)

# 12 - Filtre e exiba todas as linhas onde a altura é  maior que 1.65.
#df[df['altura']  > 1.65]               # Filtra as linhas onde a altura é maior que 1.65
#display(df[df['altura'] > 1.65])


# 13 - Modifique a altura de alguém para 1.91 e exiba o DataFrame.
df.loc[6,'altura'] = 1.91              # Modifica a altura da linha 6 para 1.91
#display(df)

# 14 - Crie uma coluna com valores de CPF para cada linha
df['CPF'] = ['123.456.789-00','987.654.321-00','111.222.333-44','555.666.777-88','999.888.777-66','000.111.222-33','444.555.666-77']
#display(df)

# 15 - Exclua a coluna CPF do DataFrame.
df.drop(columns='CPF', inplace=True)   # Exclui a coluna "CPF"
#display(df)
# 16 - Resete o índice do DataFrame e exiba o resultado.
df.reset_index(drop=True, inplace=True)  # Reseta o índice do DataFrame
#display(df)

# 17 - Exiba as linhas do DataFrame onde a altura é maior que 1.70.
df[df['altura'] > 1.70]    # Filtra as linhas onde a altura é maior que 1.70
#display(df[df['altura'] > 1.70])

# 18 - Selecione e exiba as colunas "nome" e "genero".
df[['nome', 'genero']]         # Seleciona as colunas "nome" e "genero"
#display(df[['nome', 'genero']])


# 19 - Calcule e exiba a soma das idades.
df['idade'].sum()    # Calcula a soma das idades com o método .sum()
#display(df['idade'].sum())


# 20 - Salve o DataFrame em um arquivo CSV.
df.to_csv('dados.csv', index=False)   # Salva o DataFrame em um arquivo CSV. Index=False não salva o índice do DataFrame.
display(df)





# %%
```
