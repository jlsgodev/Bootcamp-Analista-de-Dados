```python
# Instalar bibliotecas
!pip install pandas google-cloud-storage mysql-connector-python

# Importar bibliotecas
import pandas as pd
from google.cloud import storage
import mysql.connector
from mysql.connector import Error

# Autenticar a conta do Google Cloud
from google.colab import auth
auth.authenticate_user()

# Configurar o cliente do Google Cloud Storage
client = storage.Client()
bucket_name = 'bucket_sql_colab'
file_path = '/content/balanco_petro.xlsx'

# Função para fazer upload de arquivos para o bucket GCS
def upload_to_gcs(bucket_name, file_path):
    bucket = client.get_bucket(bucket_name)
    blob = bucket.blob('dados_brutos/balanco_petro.xlsx')
    blob.upload_from_filename(file_path)
    print(f'Arquivo {file_path} carregado para {bucket_name}/dados_brutos/balanco_petro.xlsx')

upload_to_gcs(bucket_name, file_path)
```

### Função para ler um arquivo Excel (.xlsx) do GCS para um DataFrame


```python
# Função para ler um arquivo Excel (.xlsx) do GCS para um DataFrame
def read_excel_from_gcs(bucket_name, file_name):
    try:
        # Obter o bucket
        bucket = client.get_bucket(bucket_name)
        # Obter o arquivo no bucket (blob)
        blob = bucket.blob(file_name)
        # Baixar o arquivo para o ambiente local
        local_path = '/content/balanco_petro.xlsx'  # Caminho local para salvar o arquivo
        blob.download_to_filename(local_path)
        # Verificar se o arquivo foi baixado
        print(f"Arquivo {local_path} baixado com sucesso!")
        # Ler o arquivo Excel com pandas
        df = pd.read_excel(local_path)
        print(f"Arquivo {local_path} lido com sucesso!")
        return df
    except Exception as e:
        print(f"Ocorreu um erro ao ler o arquivo: {e}")
        return None

# Usar a função para ler o arquivo Excel
df = read_excel_from_gcs(bucket_name, 'dados_brutos/balanco_petro.xlsx')

# Exibir as primeiras linhas do DataFrame
if df is not None:
    display(df.head())
```

### Tratar df

```python

import unicodedata

def remover_acentos(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')

def tratar_nome_coluna(nome_coluna):
    nome_coluna = remover_acentos(nome_coluna)  # Retira acentos
    nome_coluna = nome_coluna.replace(' ', '_').lower()  # Substitui espaços por '_', e deixa em minúsculas
    return nome_coluna

# Usa pandas .apply() para tratar os nomes das colunas
def renomear_colunas(df):
    colunas_tratadas = df.columns.to_series().apply(tratar_nome_coluna)
    # Evitar nomes duplicados com pandas' função `duplicated` para adicionar sufixo numérico
    colunas_unicas = colunas_tratadas.where(~colunas_tratadas.duplicated(), colunas_tratadas + '_' + colunas_tratadas.groupby(colunas_tratadas).cumcount().astype(str))
    df.columns = colunas_unicas
    return df

# Exemplo de uso
df_transposto = renomear_colunas(df_transposto)

display(df_transposto)
```
